## Question 2a

sequence: a_n = a_n−1 + a_n−2 + a_n−3           

base cases: a1 = a2 = a3 = 1

### Method 1 (recursive)

```python
#METHOD 1 (recursive)
def seq_2a(n):
    base_cases = [1, 1, 1]
    for x in range(n-1):
        next = base_cases[-1] + base_cases[-2] + base_cases[-3] #previous + n-2 + n-3
        base_cases.append(next)
    return base_cases
```

```python
#First 100 terms of sequence 2a 
print('n= ', 'an = an−1 + an−2 + an−3')
n_2ar = []
for n in range(1, 101):
    n_2ar.append(n)
    print(n, ' ', seq_2a(n)[-1]) #only print latest number in list (number relating to given n)
```

```python
#GRAPH STYLE 1 (NON-LOGARITHMIC SCALE)
plt.plot(seq_2a(100))

plt.xlabel('n-values')
plt.ylabel('sequence values')
plt.grid(True)
plt.title('Sequence 2a: a_n = a_n−1 + a_n−2 + a_n−3')
plt.figtext(0.1, -0.02 ,'This is a graph showing the values for n in sequence 2a.')
plt.show()
```

```python
#GRAPH STYLE 2 (LOGARITHMIC SCALE)
plt.plot(seq_2a(100))

plt.xlabel('n')
plt.ylabel('sequence values on a logarithmic scale')
plt.grid(True)
plt.yscale('log')
plt.title('Sequence 2a: a_n = a_n−1 + a_n−2 + a_n−3')
plt.figtext(0.1, -0.02 ,'This is a graph showing the values for n in sequence 2a on a logarithmic y-scale.')
plt.show()
```

### Method 2 (non-recursive)

```python
#METHOD 2 (non-recursive)
def sequence_2a(n):
    if n <= 0: 
        return 'n/a'
    elif n == 1:# or 2 or 3:
        return 1
    elif n == 2:
       return 1
    elif n == 3:
       return 1
    else:
        a, b, c = 1, 1, 1 #note: here a = an-1, b = an-2, c = an-3
        for x in range(n-1):                                        
            #add previous 3 numbers in sequence, a+b+c
            
            #a, b, c = b, c, a + b + c - works but have to return a
            #c, b, a = a + b + c, c, b - no
            a, b, c = a+b+c, a, b
            #a now = prev answer
            #b now = a
            #c now = b
        
            #c, b, a = b, a, a + b + c # - also correct
        return c #a_n for this n
```

## Question 2b

sequence: a_n = a_n−1 − 2a_n−2

base cases: a1 = a2 = 1

```python
def sequence_2b(n):
    if n <= 0: 
        return 'n/a'
    elif n == 1:
        return 1
    elif n == 2:
       return 1
    else:
        a, b = 1, 1
        for x in range(n-1):                                         
            #subtract (2x -2previous) from previous
            a, b = a - (2*b), a
        return b
```

```python
#Test
sequence_2b(5) == -1 #True - calculations correct
```

```python
#First 100 terms of sequence 2b
print('n= ', 'an = an−1 - 2an−2')
sequence_2b_nums = []
n_2b = []
for n in range(1, 101):
    print(n, ' ', sequence_2b(n))
    sequence_2b_nums.append(sequence_2b(n)) #creating a list of first 100 terms in the sequence
    n_2b.append(n)
```

## Question 2c

sequence: a_n = 4a_n−1(1 − a_n−1)

base cases: a1 = 0.2

```python
def sequence_2c(n):
    if n <= 0: 
        return 'n/a'
    elif n == 1:
        return 0.2
    else:
        a = 0.2
        for x in range(n-1):                                         
            #subtract  number (-2) from previous
            a = (4*a) * (1-a)
        return a
```

## Question 3

#### Plot in Python the first 50 terms of the sequence a_n = a_n−1 + a_n−2 with initial conditions a_1 = 0, a_2 = 1.

iterative rule: a_n = a_n−1 + a_n−2

base cases: a_1 = 0, a_2 = 1

```python
def sequence_3(n):
    if n <= 0:
        return 'n/a'
    elif n == 1:
        return 0
    elif n == 2:
       return 1
    else:
        a, b = 1, 0 #note: here a = an-1, b = an-2        e.g. a3 = a2 + a1 therefore a3 = 1 + 0 (or a + b)
        for x in range(n-1):                                       
            #add previous 2 numbers in sequence, a+b
            a, b = a+b, a
            #a now = prev answer
            #b now = prev a
        return b #a_n for this n
```

```python
#First 50 terms of sequence 3
print('n: ', 'a_n = a_n−1 + a_n−2')
sequence_3_nums= []
for n in range(1, 51):
    print(n, ' ', sequence_3(n))
    sequence_3_nums.append(sequence_3(n)) #creating a list of first 50 terms in the sequence
```
